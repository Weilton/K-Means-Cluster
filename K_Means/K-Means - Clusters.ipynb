# Identifying Clusters values

```python
# imports
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pylab as pl
```

```python
# Load dataset
iris = load_iris()
iris
```

```python
# Visualize the type
type(iris.data)
```

```python
# Create a model with 3 clusters
kmeans = KMeans(n_clusters = 3)
kmeans.fit(iris.data)
```

```python
# Visualize the clusters created
labels = kmeans.labels_
labels
```

```python
# Converte the data to a dataframe and create a cluster Map
cluster_map = pd.DataFrame(iris.data)

# Add a column 'cluster' and to assign the labels
cluster_map['cluster'] = labels
cluster_map
```

```python
# We can get only one cluster
cluster_map[cluster_map.cluster == 2]
```

# PCA - Reduce Dimensionality

```python
pca = PCA(n_components = 2).fit(iris.data)
```

```python
# Applying the PCA
pca_2d = pca.transform(iris.data)
pca_2d
```

```python
pca_2d.shape
```

```python
pca_2d.shape[0]
```

```python
# Lets to plot
for i in range(0, pca_2d.shape[0]):
    if labels[i] == 0:
        c1 = pl.scatter(pca_2d[i,0], pca_2d[i,1], c = 'r', marker = '+')
        
    elif labels[i] == 1:
        c2 = pl.scatter(pca_2d[i,0], pca_2d[i,1], c = 'g', marker = 'o')
        
    elif labels[i] == 2:
        c3 = pl.scatter(pca_2d[i,0], pca_2d[i,1], c = 'b', marker = '*')
        pl.legend([c1,c2,c3], ['Cluster 0', 'Cluster 1', 'Cluster 2'])
        pl.title('K-Means with dataset iris and 3 clusters')
pl.show()
```
